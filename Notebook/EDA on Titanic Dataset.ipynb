```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\ranik\\Downloads\\Titanic-Train-Data.csv')
data.head()
```

```python
data.columns
```

```python
data.isna().sum()
```

```python
data['Age'].median()
```

```python
data['Age']=data['Age'].fillna(value=28)
data['Age'].isna().sum()
```

```python
data['Cabin'].fillna(value=0,inplace=True)
```

```python
# Ticket 1st letters
for k in data.Ticket:
    print(k[0])
```

```python
data['Embarked'].value_counts()
```

```python
surv=data.groupby('Survived')
surv['Embarked'].value_counts()
```

```python
data['Embarked'] = data['Embarked'].fillna(value='S')
data.head()
```

```python
# Converting male as 1 and female as 0 for Age
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

data['Sex'] = label_encoder.fit_transform(data['Sex'])
```

```python
data['Sex'][:5]
```

```python
data['Sex'].value_counts()
```

```python
# Converting S as 2 , C as 1 , Q as 0 in Embarked
data['Embarked'] = label_encoder.fit_transform(data['Embarked'])
```

```python
data['Embarked'][:5]
```

```python
data['Embarked'].value_counts()
```

```python
# Adding parents , childrens and sbilings
data['family'] = data['SibSp'] + data['Parch']+1
```

```python
data = data.drop(['SibSp','Parch','Embarked','PassengerId','Fare'],axis=1)
data.head()
```

```python
# Nan values heatmap
sns.heatmap(data.isna())
```
